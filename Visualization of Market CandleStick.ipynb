```python
!pip install --upgrade mplfinance
import mplfinance as mpf
import matplotlib.pyplot as plt
def getHourlyData(symbol):
  frame = pd.DataFrame(client.get_historical_klines(symbol,'1month','25 hours ago UTC'))
  frame = frame.iloc[:,:6]
  frame.columns = ['Time','Open','High','Low','Close','Volume']
  frame[['Open','High','Low','Close','Volume']] = frame[['Open','High','Low','Close','Volume']].astype(float)
  frame.Time = pd.to_datetime(frame.Time,unit='ms')
  frame.set_index(['Time'],inplace=True)
  return frame
#klines = getHourlyData('BTCUSDT')
frame = pd.DataFrame(client.get_historical_klines('BTCUSDT','1w','1 Mar 2020'))#,'25 hours ago UTC'))
frame = frame.iloc[:,:6]
frame.columns = ['Time','Open','High','Low','Close','Volume']
frame[['Open','High','Low','Close','Volume']] = frame[['Open','High','Low','Close','Volume']].astype(float)
frame.Time = pd.to_datetime(frame.Time,unit='ms')
frame.set_index(['Time'],inplace=True)
mpf.plot(frame ,type = 'candle', style = 'blueskies' ,tight_layout=True,title='BTCUSDT', mav = (5,8,20),volume = True)
```
